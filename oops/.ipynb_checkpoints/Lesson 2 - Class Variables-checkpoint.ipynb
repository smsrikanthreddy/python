In this notebook, we are going to learn class variables importance. How to access class variables, different ways of using class variables. They are shared among all instances of the class

Lets learn class variables using pay raise variable

##### Step 1

```python
class Employee:
    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.email = first_name+"."+last_name+"@company.com"
        self.salary = salary
        
    def fullname(self):
        return "{} {}".format(self.first_name, self.last_name)
    
    def apply_raise(self):
        self.salary = int(self.salary * 1.04)
```

```python
emp1 = Employee("Srikanth","Metlakunta",5000)
emp2 = Employee("Ramesh","Gummi",6000)

print(emp1.salary)
emp1.apply_raise()
print(emp1.salary)
```

##### Step 2

```python
class Employee:
    raise_amount = 1.04
    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.email = first_name+"."+last_name+"@company.com"
        self.salary = salary
        
    def fullname(self):
        return "{} {}".format(self.first_name, self.last_name)
    
    def apply_raise(self):
        # self.salary = int(self.salary * Employee.raise_amount)
        self.salary = int(self.salary * self.raise_amount)
        
emp1 = Employee("Srikanth","Metlakunta",5000)
emp2 = Employee("Ramesh","Gummi",6000)

print(emp1.salary)
emp1.apply_raise()
print(emp1.salary)
```

```python
print(Employee.raise_amount)
print(emp1.raise_amount) #accessing the class instance variable
print(emp1.raise_amount) #accessing the class instance variable
```

```python
Employee.__dict__
```

You can see in the below statements, we can change the class variable and the effect on class instances

```python
Employee.raise_amount = 1.05
print(Employee.raise_amount)
print(emp1.raise_amount) #accessing the class instance variable
print(emp1.raise_amount) #accessing the class instance variable
```

What if I want to change the raise amount class variable using instance instead of class

```python
emp1.raise_amount = 1.05
print(Employee.raise_amount)
print(emp1.raise_amount) #accessing the class instance variable
print(emp2.raise_amount) #accessing the class instance variable
```

to check the count of number of employees

```python
class Employee:
    raise_amount = 1.04
    num_of_emp = 0
    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.email = first_name+"."+last_name+"@company.com"
        self.salary = salary
        Employee.num_of_emp += 1
        
    def fullname(self):
        return "{} {}".format(self.first_name, self.last_name)
    
    def apply_raise(self):
        # self.salary = int(self.salary * Employee.raise_amount)
        self.salary = int(self.salary * self.raise_amount)

print(Employee.num_of_emp)
emp1 = Employee("Srikanth","Metlakunta",5000)
emp2 = Employee("Ramesh","Gummi",6000)
print(Employee.num_of_emp)
```
