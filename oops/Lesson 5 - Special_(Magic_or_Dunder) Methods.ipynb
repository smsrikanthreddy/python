In this notebook, we are going to learn about special methods also called as magic methods. Special methods is also used to implement operator overloading. special methods are often surrounded by "_  _" called dunders. if we say duner initi, that means "__ init __". By defining special methods we should be able to change the built-in behavior and operations.

```python
class Employee:

    raise_amt = 1.04

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.email = first + '.' + last + '@email.com'
        self.pay = pay

    def fullname(self):
        return '{} {}'.format(self.first, self.last)

    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amt)

    def __repr__(self):
        return "Employee('{}', '{}', {})".format(self.first, self.last, self.pay)

    def __str__(self):
        return '{} - {}'.format(self.fullname(), self.email)

    def __add__(self, other):
        return self.pay + other.pay

    def __len__(self):
        return len(self.fullname())
```

```python
emp_1 = Employee('Corey', 'Schafer', 50000)
emp_2 = Employee('Test', 'Employee', 60000)

print(emp_1)
```

```python
print(repr(emp_1))
print(str(emp_1))
```

```python
print(emp_1.__repr__())
print(emp_1.__str__())
```

```python
print(emp_1 + emp_2)
```

```python
print(len(emp_1))
```
